Valor médio, mínimo e máximo da pontuação de Matemática(math score)


```python
mediamat = df['math score'].mean()
print(" a média das notas de matemática é: ", mediamat)
```

```python
minmat = df['math score'].min()
print(" O valor mínimo das notas de matemática é: ", minmat)
```

```python
maxmat = df['math score'].max()
print("O valor máximo das notas de matemática é: ", maxmat)
```

Valor médio, mínimo e máximo da pontuação de Leitura(reading score)

```python
mediaread = df['reading score'].mean()
print("A média das notas de leitura é:", mediaread)
```

```python
minread = df['reading score'].min()
print("O mínimo das notas de leitura é:", minread)
```

```python
maxread = df['reading score'].max()
print("A média das notas de leitura é:", maxread)
```

Valor médio, mínimo e máximo da pontuação de Escrita(writing score)

- Agrupe o banco de dados a partir da raça/etnia dos alunos, e
plote um gráfico com a nota média em Matemática para cada
grupo disponível;

```python
df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2.plot(kind='bar')
```

- Agrupe o banco de dados a partir do Nível de Escolaridade dos
pais, e plote um gráfico com a nota média em Escrita para cada
grupo disponível;

```python
df3 = df.groupby('parental level of education')[['writing score']].mean()
df3.plot(kind='bar')
```

Agrupe o banco de dados a partir da coluna referente à participação
em um curso preparatório para os testes, e plote um gráfico com a
nota média em Leitura para cada grupo disponível.

```python
df4 = df.groupby('test preparation course')[['reading score']].mean()
df4.plot(kind='bar', color='red')
```

```python
import seaborn as sns

# Exemplo de pair plot
sns.pairplot(df[['math score', 'reading score', 'writing score']], diag_kind='kde')
plt.show()
```

A média geral das disciplinas está na faixa dos 60 a 70 pontos, o que pode indicar que, embora muitos alunos estejam obtendo notas razoáveis, há um número significativo de alunos que está abaixo da média ideal, puxando a média geral para baixo.
